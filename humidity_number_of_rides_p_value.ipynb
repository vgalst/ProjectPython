```python

final_df = pd.DataFrame({"Number Of Rides":count_data,
                       "Temperature":avg_temp,
                       "Wind_Speed":avg_wind,
                       "Humidity":avg_humidity,
                       })
final_df
```

```python
# calculate the time of the rain
final_df = final_df.reset_index()
final_df.head()
```

```python
sample_rides_humidity = final_df[["Pickup_Datetime","Number Of Rides","Humidity","Weather_Description"]]
sample_rides_humidity.head()
```

```python
number_of_rides_mean = sample_rides_humidity["Number Of Rides"].mean()
number_of_rides_mean
```

```python
humidity_range = sample_rides_humidity["Humidity"].unique().tolist()
humidity_range
```

```python
humidity_low = sample_rides_humidity.loc[sample_rides_humidity["Humidity"] < 20]
humidity_low
```

```python
number_of_rides_lowhum = humidity_low["Number Of Rides"].mean()
number_of_rides_lowhum
```

```python
humidity_normal = sample_rides_humidity.loc[sample_rides_humidity["Humidity"]>20]
humidity_normal = humidity_normal.loc[humidity_normal["Humidity"]<60]
humidity_normal
```

```python
number_of_rides_normalhum = humidity_normal["Number Of Rides"].mean()
number_of_rides_normalhum
```

```python
humidity_high = sample_rides_humidity.loc[sample_rides_humidity["Humidity"]>75]
humidity_high
```

```python
number_of_rides_highhum = humidity_high["Number Of Rides"].mean()
number_of_rides_highhum
```

```python
print(f'The average number of rides in uncomforatable dry humidity is {number_of_rides_lowhum}.')
```

```python
print(f'The average number of rides in comfort range humidity is {number_of_rides_normalhum}.')
```

```python
print(f'The average number of rides in uncomforatable humidity is {number_of_rides_highhum}.')
```

```python
expected_value = (number_of_rides_lowhum + number_of_rides_normalhum + number_of_rides_highhum) / 3
expected_value
```

```python
observed = pd.Series([number_of_rides_lowhum,number_of_rides_normalhum,number_of_rides_highhum], index=["Uncomforatable Dry", "Comfort Range", "Uncomforatable Wet"])
observed
```

```python
df = pd.DataFrame([observed]).T
df.head()
```

```python
df[1] = expected_value
```

```python
df.columns = ["Observed", "Expected"]
df
```

```python
# The degree of freedom is 3-1 = 2
# With a p-value of 0.05, the confidence level is 1.00-0.05 = 0.95.
critical_value = stats.chi2.ppf(q = 0.95, df = 2)
critical_value
```

```python
# Run the chi square test with stats.chisquare()
stats.chisquare(df['Observed'], df['Expected'])
```
