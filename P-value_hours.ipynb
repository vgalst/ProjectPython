```python
file_data =pd.DataFrame(Yearly_Data)
file_data
```

```python

sort_file = file_data.rename(columns = {"Pickup_DateTime" : "Pickup_Datetime",
                                       "DropOff_datetime" : "Dropoff_Datetime",
                                       "Pickup_ZipCode" : "Pickup_Zipcode",
                                       "Dropoff_ZipCode" : "Dropoff_Zipcode",
                                       "Temp":"Temperature",
                                       "humidity":"Humidity",
                                       "wind_speed":"Wind_Speed",
                                       "weather_id":"ID_Weather",
                                       "weather_main" : "Weather_Like",
                                       "weather_description" : "Weather_Description"})
sort_file.head()



sort_file["Trip_Duration"] = sort_file["Trip_Duration"].astype(float)
sort_file


sort_file.sort_values(["Pickup_Datetime"], ascending = True)
```

```python
sort_file["Pickup_Datetime"] = sort_file["Pickup_Datetime"].astype('datetime64[ns]')
sort_file['Pickup_Datetime'] = sort_file['Pickup_Datetime'].apply(lambda x: x.replace(minute=0, second=0))
```

```python
new = sort_file['Pickup_City'].str.split(',',n=1,expand = True)
sort_file['Borough'] = new[1]
```

```python
sort_file
```

```python
sort_file["hours"] =pd.to_datetime(sort_file["Pickup_Datetime"]).dt.hour
sort_file["hours"]


sort_file=sort_file.assign(hours_group=pd.cut(sort_file.Pickup_Datetime.dt.hour,
                          [-1,6,12,18,23],
                          labels=['Night','Morning','Afternoon','Evening'],
                          include_lowest=True))

sort_file.head()
```

```python
p_value_data = sort_file[["Trip_Duration","hours_group",'hours']]
p_value_data





```

```python
gb = p_value_data.groupby('hours_group').mean().reset_index()
gb.head()

```

```python
gb.dtypes
night_hr= gb.iloc[0]['hours']
night_hr
```

```python
morning_hr = gb.iloc[1]['hours']
morning_hr
```

```python
afternoon_hr = gb.iloc[2]['hours']
afternoon_hr
```

```python
evening_hr = gb.iloc[3]['hours']
evening_hr
```

```python
observed = pd.Series([night_hr,morning_hr,afternoon_hr,evening_hr], index=["night", "morning", "afternoon","evening"])
observed
```

```python
df = pd.DataFrame([observed]).T
df.head()
```

```python
expected_value = (night_hr + morning_hr + afternoon_hr + evening_hr) / 4
expected_value
```

```python
df[1] = expected_value
```

```python
df.columns = ["Observed", "Expected"]
df
```

```python
# The degree of freedom is 4-1 = 3
# With a p-value of 0.05, the confidence level is 1.00-0.05 = 0.95.
critical_value = stats.chi2.ppf(q = 0.95, df = 3)
critical_value
```

```python
# Run the chi square test with stats.chisquare()
stats.chisquare(df['Observed'], df['Expected'])
```
