```python
# Converting Kelvin To Fahrenheit 
def kelvin_to_fahrenheit(tempereture_kelvin):
    tempereture_fahrenheit = (((tempereture_kelvin - 273) * 9/5) + 32)
    return tempereture_fahrenheit
```

```python
nyc_temp = rideshare_df[["Pickup_DateTime",
                                "DropOff_datetime",
                                "temp"]]
nyc_temp.head()
```

```python

# rename the temp to Tempereture F
nyc_temp_f = nyc_temp.rename(columns = {"temp" :"Tempereture F"})
nyc_temp_f.head()
nyc_temp_f["Tempereture F"] = kelvin_to_fahrenheit(nyc_temp["Tempereture F"])
nyc_temp_f.head()
```
