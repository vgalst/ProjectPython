```python

final_df = pd.DataFrame({"Number Of Rides":count_data,
                       "Temperature":avg_temp,
                       "Wind_Speed":avg_wind,
                       "Humidity":avg_humidity,
                       })
final_df
```

```python
# calculate the time of the rain
final_df = final_df.reset_index()
final_df.head()
```

```python
sample_rides_temperature = final_df[["Pickup_Datetime","Number Of Rides","Temperature","Weather_Description"]]
sample_rides_temperature.head()
```

```python
number_of_rides_mean = sample_rides_temperature["Number Of Rides"].mean()
number_of_rides_mean
```

```python
temperature_unique = sample_rides_temperature["Temperature"].unique().tolist()
temperature_unique
```

```python
temperature_cold = sample_rides_temperature.loc[sample_rides_temperature["Temperature"] < 60]
temperature_cold
```

```python
number_of_rides_cold = temperature_cold["Number Of Rides"].mean()
number_of_rides_cold
```

```python
temperature_normal = sample_rides_temperature.loc[sample_rides_temperature["Temperature"]>60]
temperature_normal = temperature_normal.loc[temperature_normal["Temperature"]<75]
temperature_normal
```

```python
number_of_rides_normal = temperature_normal["Number Of Rides"].mean()
number_of_rides_normal
```

```python
temperature_hot = sample_rides_temperature.loc[sample_rides_temperature["Temperature"]>75]
temperature_hot
```

```python
number_of_rides_hot = temperature_hot["Number Of Rides"].mean()
number_of_rides_hot
```

```python
print(f'The average number of rides in cold temperature is {number_of_rides_cold}.')
```

```python
print(f'The average number of rides in moderate temperature is {number_of_rides_normal}.')
```

```python
print(f'The average number of rides in hot temperature is {number_of_rides_hot}.')
```

```python
expected_value = (number_of_rides_cold + number_of_rides_normal + number_of_rides_hot) / 3
expected_value
```

```python
observed = pd.Series([number_of_rides_cold,number_of_rides_normal,number_of_rides_hot], index=["Cold", "Moderate", "Hot"])
observed
```

```python
df = pd.DataFrame([observed]).T
df.head()
```

```python
df[1] = expected_value
```

```python
df.columns = ["Observed", "Expected"]
df
```

```python
# The degree of freedom is 3-1 = 2
# With a p-value of 0.05, the confidence level is 1.00-0.05 = 0.95.
critical_value = stats.chi2.ppf(q = 0.95, df = 2)
critical_value
```

```python
# Run the chi square test with stats.chisquare()
stats.chisquare(df['Observed'], df['Expected'])
```
