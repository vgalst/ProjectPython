```python
# Dependencies
import openweathermapy.core as owm
# Dependencies
import requests
import random
#from config import api_key
from config import gkey
import pandas as pd
import matplotlib.pyplot as plt
```

```python
location_file = "fhv_zone_lookup.csv"

src_df = pd.read_csv(location_file)
location_df =src_df.dropna()
location_df
```

```python
target_city = location_df["Borough"]+', '+location_df["Zone"]
#target_city = location_df["Borough"]+', '+location_df["Zone"]+', '+location_df["service_zone"]

target_city
```

```python
zipcode = []
for city in target_city:
    target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'                     
              'address={0}&key={1}').format(city, gkey)
    geo_data = requests.get(target_url).json()
    lat = geo_data["results"][0]["geometry"]["location"]["lat"]
    lng = geo_data["results"][0]["geometry"]["location"]["lng"]
    try : 
        zc = geo_data["results"][0]["address_components"][6]['long_name']
        #zipcode.append(zc)
        
        if zc =='New York':
            zc = geo_data["results"][0]["address_components"][8]['long_name']
            zipcode.append(zc)
        elif zc =='United States':
            zc = geo_data["results"][0]["address_components"][7]['long_name']
            zipcode.append(zc)
            
        else:
            zipcode.append(zc)
            
            
            
        
    except IndexError: 
        
         #print(lat, lng)
        ziptarget_url = ('https://maps.googleapis.com/maps/api/geocode/json?latlng=40.714224,-73.961452&key={0}').format (gkey)
        georev_data = requests.get(ziptarget_url).json()
        zc =georev_data["results"][0]["address_components"][7]['long_name']
        zipcode.append(zc)

zipcode        

    
    
```

```python
location_df['zipcode'] = zipcode
location_df
```

```python
location_df.to_csv("fhv_zone_lookup_zipcode.csv", index=False)
```
